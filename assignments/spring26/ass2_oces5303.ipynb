See the samples assignments I've made for the kind of things we might be expecting. We will probably be fairly loose with giving credit. 50% or below would count as unsatisfactory, while around 85% or above would be an A grade.

You are allowed to use other Python packages, but see point b) below.

If you upload your assignment onto a GitHub repository and give us the link, with the appropriate things built in so that we can open it in Colab, then you get extra credit because of making the things openly reproducible and helping us somewhat (the things still need to run from fresh; just putting it on GitHub and linking to Colab by itself is worth nothing if it doesn't run).

c) ***Plagiarism***: By all means consult each other, work together, or use AI tools, but the files you hand in should be done and written up separately. For allowing checks in Turnitin, you should pass it through `Kernel -> Restarts & Clear Output` before you hand it in. **The default for anyone accused with plagiarism is ZERO on the assignment**, and depending on whether you decide to contest and the result of the appeal, possibly lead to an official note of plagiarism on your transcript (I will allow people to argue but be be ready for possible consequences). 

A few things count as plagiarism:

**Copying between students, and the default is that ALL parties involved get zero for the assignment**, regardless of whether the side can demonstrate they were copied from (extra incentive to keep the writing separate).

**Copying text without citation is plagiarism**. Use quotation marks and give reference if you are directly lifting text, but don't do this too often (and will result in text looking cluttered, and not getting full credit for the *presentation* aspect). This includes content generated from AI.

**Code is a slightly more grey area**, but I will just say no one has ever really been punished for being cautious and generous with citations, but make sure you present it well (e.g. overburdening text with citations will make the presentation ugly, and will not get full credit for the *presentation* aspect say). This includes content generated from AI.

I will just make the point that we don't tend to accuse plagiarism unless we have enough proof, and if we are doing it it probably means we think we have a sufficiently strong case that is probably not worth arguing against (because then penalty then gets increased).
