For allowing checks in Turnitin, you should pass it through `Kernel -> Restarts & Clear Output` before you hand it in. **The default for anyone accused with plagiarism is ZERO on the assignment**, and depending on whether you decide to contest and the result of the appeal, possibly lead to an official note of plagiarism on your transcript (I will allow people to argue but be be ready for possible consequences). 

A few things count as plagiarism:

**Copying between students, and the default is that ALL parties involved get zero for the assignment**, regardless of whether the side can demonstrate they were copied from (extra incentive to keep the writing separate).

**Copying text without citation is plagiarism**. Use quotation marks and give reference if you are directly lifting text, but don't do this too often (and will result in text looking cluttered, and not getting full credit for the *presentation* aspect). This includes content generated from AI.

**Code is a slightly more grey area**, but I will just say no one has ever really been punished for being cautious and generous with citations, but make sure you present it well (e.g. overburdening text with citations will make the presentation ugly, and will not get full credit for the *presentation* aspect say). This includes content generated from AI.

I will just make the point that we don't tend to accuse plagiarism unless we have enough proof, and if we are doing it it probably means we think we have a sufficiently strong case that is probably not worth arguing against (because then penalty then gets increased).

# Extend project (25% of total course grade)

For this you should use your own data and deploy your choice of machine learning things to investigate your chocie of data; you can do any number of tools (and you may want to do at least two), but one of these have to be a neural network of some description.

For this one we want a submission as an qualifying exam report document (or an article document if you like) in pdf format. We don't need to see the coding details (you can show it if you think it helps the narrative). You need to describe what the data is, what is the scientific question you want to address, the details of the machine learning algorithm you deploy and why that is a suitable choice, the choice of model parameters/architecture and robustness to those, a critique/evaluation of the results and a conclusion. You would also need to defend your choices in the interview style exam (cf. qualifying exam Q&A session).

***Note that it doesn't need to work*** ("work" in whatever sense you want it to mean). The interest here is not in getting something that works, but to demonstrate you can deploy things, speculate why it doesn't work, identify how you might make it work, and talk about those. There is no one way that is "right" (just like there is no single way of something being a "good" article, although there are usually commonalities between "good" things). I would recommend doing two flavours of neural networks, doing one single task (classification or regression), and being reasonably thorough with model comparison, sensitivity on hyper-parameter exploration, model validation and so forth.

<u>Note</u>: For the present purpose, you are allowed to cite Wikipedia, but you probably also want to not have Wikipedia be your only source.

***You should name your hand-in file as something like "project_oces5303_STUDENTID.pdf" when you hand the notebook in through Canvas***.
